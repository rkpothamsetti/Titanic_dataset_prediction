```python
import pandas as pd
import numpy as np

df = pd.read_csv("/content/drive/MyDrive/Titanic-Dataset.csv")

# Summary statistics
summary_stats = df.describe(include=[np.number])
summary_stats.loc["median"] = df.median(numeric_only=True)
print(summary_stats)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

sns.set(style="whitegrid")
numeric_cols = ['Age', 'Fare', 'SibSp', 'Parch', 'Pclass']

fig, axs = plt.subplots(len(numeric_cols), 2, figsize=(12, 18))
for i, col in enumerate(numeric_cols):
    sns.histplot(df[col].dropna(), bins=30, ax=axs[i, 0], kde=True, color='skyblue')
    axs[i, 0].set_title(f'Histogram of {col}')

    sns.boxplot(x=df[col], ax=axs[i, 1], color='lightgreen')
    axs[i, 1].set_title(f'Boxplot of {col}')

plt.tight_layout()
plt.show()
```

```python
# Correlation matrix
corr_matrix = df.corr(numeric_only=True)
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm')
plt.title("Correlation Matrix")
plt.show()


```

```python
# Pairplot
sns.pairplot(df[['Age', 'Fare', 'SibSp', 'Parch', 'Pclass', 'Survived']].dropna(), hue='Survived')
plt.show()
```
